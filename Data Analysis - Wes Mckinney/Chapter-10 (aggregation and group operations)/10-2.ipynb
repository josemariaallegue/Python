```python
import pandas as pd
from pathlib import Path


def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]


df = pd.read_csv(Path(r"D:\Documentos\GitHub\Python\Data Analysis - Wes Mckinney\Archivos\tips.csv"),
                 sep=",", encoding="utf8", engine="c", header=0)
df
```

```python
df['tip_pct'] = df['tip'] / (df['total_bill'] - df['tip'])
# top(df)
# se aplica la funcion sobre los los grupos generados.
# en este caso "smoker-yes" y "smoker-no"
df.groupby(["smoker"]).apply(top)
```

Se puede quitar la creacion de los group keys facilmente

```python
df.groupby("smoker", group_keys=False).apply(top)
```

```python
df.pivot_table(['tip_pct', 'size'], index=['time', 'day'],
               columns='smoker', margins=True, aggfunc="count")
```
