Se pueden hacer plots desde dataframes o series

```python
import numpy as np
import pandas as pd
from pathlib import Path
import seaborn as sns

df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))

df.plot.bar()

# alpha permite cambiar la tonalidad
df.plot.barh(alpha=0.5)
```

```python
tips = pd.read_csv(Path(r"D:\Documentos\GitHub\Python\Data Analysis - Wes Mckinney\Chapter-9 (Plotting and Visualization)\tips.csv"),
                   sep=",", encoding="utf8", header=0)
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])

tips.head()
```

Para hacer plot con seaborn solo se lo llama, se busca la funcion necesario para el grafico y se pasan los datos

```python
sns.barplot(x="tip_pct", y="day", data=tips, orient="h")
```

```python
sns.barplot(x="tip_pct", y="day", hue="time", data=tips, orient="h")
```

Formas de hacer un grafico de distribucion

```python
tips["tip_pct"].plot.hist(bins=50)
sns.histplot(x="tip_pct", data=tips)
```

```python
tips["tip_pct"].plot.density()
sns.displot(tips["tip_pct"], color="k", kde=True)
```

```python
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker',
            kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x='tip_pct', y='day', kind='box',
               data=tips[tips.tip_pct < 0.5])
```
