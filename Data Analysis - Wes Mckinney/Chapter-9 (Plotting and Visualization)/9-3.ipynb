```python
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

# este permite determinar la evolucion del rango del axis "x"
ticks = ax.set_xticks([0, 250, 500, 750, 1000])

# este permite determinar el nombre del rango del axis "x"
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')

# setea el titulo del plot
ax.set_title("My first matplotlib plot")

# setea el titulo del axis
ax.set_xlabel("Stages")

# tambien se puede setear todos a travez de un diccionario
props = {
 'title': 'My first matplotlib plot',
 'xlabel': 'Stages'
}
ax.set(**props)
```

```python
# para identificar los plots facilmente se puede asignarle un label durante la creacion
fig2 = plt.figure()
ax = fig2.add_subplot(1,1,1)
ax.plot(randn(50).cumsum(), "k", label="one")
ax.plot(randn(50).cumsum(), "k--", label="two")
ax.plot(randn(50).cumsum(), "k.", label="three")

ax.legend(loc="best")
```
